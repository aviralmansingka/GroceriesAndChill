```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./transaction_data.csv')
product_data = pd.read_csv('./product.csv')

data = data.merge(product_data, on='PRODUCT_ID')
data
```

```python
data = data.ix[1:200000, ['BASKET_ID', 'COMMODITY_DESC', 'QUANTITY']]
data.head()
data.shape
```

```python
table = pd.pivot_table(data, values='QUANTITY', index=['BASKET_ID'], columns=['COMMODITY_DESC'])
```

```python
table = table.fillna(value=0)
table.shape
```

```python
table.head()
```

```python
shopping_list = np.zeros((100,), dtype=np.int)

shopping_list[77] = 3
shopping_list[21] = 4
shopping_list[11] = 5

my_data = pd.Series(shopping_list).ix[:]
```

```python
non_zero = [index for (index, item) in enumerate(shopping_list) if item > 0]
non_zero
```

```python
entries = [table.loc[table.ix[:,index] > 0] for index in non_zero]

len(entries)
entries = pd.concat(entries)
```

```python
cosine_data = entries.ix[:,:].as_matrix()
```

```python
from scipy import spatial

my_data = pd.Series(shopping_list).ix[:]
similarity_list = [(index, 1 - spatial.distance.cosine(my_data, item)) for (index, item) in enumerate(cosine_data)]
```

```python
similarity_list.sort(key=lambda x: x[1], reverse=True)[:10]
similarity_list
```

```python
similarity_rows = [row for (row, similarity) in similarity_list]
similarity_rows.sort()
```

```python
similarity_data = table.iloc[similarity_rows, :].as_matrix()
summed_similarities = np.sum(similarity_data, axis=0)

item_pairs = [(index, value) for (index,value) in enumerate(summed_similarities)]
item_pairs.sort(key=lambda x: x[1], reverse=True)
item_pairs = item_pairs[:3]

column_indices = [column for (column, score) in item_pairs]
column_indices

my_list = [get_col_name(index) for index in column_indices]
```

```python
def get_col_name(index):
    return table.columns.values[index]
    
def getRecommendations(shopping_list, table, k):
    """function takes as input the shopping list categories and recommends to the users new items"""
    
    non_zero = [index for (index, item) in enumerate(shopping_list) if item > 0]

    entries = [table.loc[table.ix[:,index] > 0] for index in non_zero]
    entries = pd.concat(entries)
    
    cosine_data = entries.ix[:,:].as_matrix()
    
    similarity_list = [(index, 1 - spatial.distance.cosine(my_data, item)) for (index, item) in enumerate(cosine_data)]
    similarity_list.sort(key=lambda x: x[1], reverse=True)
    similarity_list = similarity_list[:10]
    
    similarity_rows = [row for (row, similarity) in similarity_list]
    similarity_rows.sort()
    
    similarity_data = table.iloc[similarity_rows, :].as_matrix()
    summed_similarities = np.sum(similarity_data, axis=0)

    item_pairs = [(index, value) for (index,value) in enumerate(summed_similarities)]
    item_pairs.sort(key=lambda x: x[1], reverse=True)
    item_pairs = item_pairs[:k]
    
    return [get_col_name(index) for (index, score) in item_pairs]

def print_shopping_list(shopping_list):
    print_list = [get_col_name(item) for item in shopping_list if item > 0]
    return print_list
```

```python
getRecommendations(shopping_list, table, 3)
```
